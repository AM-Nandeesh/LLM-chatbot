```python
import pandas as pd

with open('/human_chat.txt', 'r') as file:
    lines = file.readlines()

data = pd.DataFrame(columns=['Human 1', 'Human 2'])


for i in range(0, len(lines)-1, 2):
    human1 = lines[i].strip().replace('Human 1:', '')
    human2 = lines[i+1].strip().replace('Human 2:', '')
    row = {'Human 1': human1, 'Human 2': human2}
    data = pd.concat([data, pd.DataFrame([row])], ignore_index=True)

if len(lines) % 2 != 0:
    human1_last = lines[-1].strip().replace('Human 1:', '')
    row = {'Human 1': human1_last, 'Human 2': ''}
    data = pd.concat([data, pd.DataFrame([row])], ignore_index=True)
```

```python
df = data.rename(columns={"Human 2":"target_text", "Human 1":"source_text"})
df = df[['source_text', 'target_text']]
```

```python
df.to_csv('chats.csv', index=False)
```

```python
df = pd.read_csv('/content/chats.csv')
df.head(-1)
```

```python
import re

def clean_text(text):
    if not isinstance(text, str):
        return text

    # Remove emojis
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  
        u"\U0001F300-\U0001F5FF"  
        u"\U0001F680-\U0001F6FF"  
        u"\U0001F1E0-\U0001F1FF"  
                           "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove other unwanted characters
    text = re.sub(r'[^\w\s]', '', text)

    return text



df['source_text'] = df['source_text'].apply(clean_text)
df['target_text'] = df['target_text'].apply(clean_text)
```

```python
df['source_text'] = "chat: " + df['source_text']
df = df.applymap(str)
df.head(-1)
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.3,random_state=101,shuffle=True)
train_df.shape, test_df.shape
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
model.train(train_df=train_df,
            eval_df=test_df, 
            source_max_token_len=128, 
            target_max_token_len=128, 
            batch_size=16, max_epochs=10, use_gpu=True,early_stopping_patience_epochs=0,save_only_last_epoch=True)
```
